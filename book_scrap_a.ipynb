```python
name = []
price = []
rating = []
avilablity = []
book_info_link = []

for i in range(1, 51):
    try:
        url = f"https://books.toscrape.com/catalogue/page-{i}.html"
        html_text = requests.get(url, timeout=10).text
        soup = BeautifulSoup(html_text, "lxml")
        contants = soup.find_all('article', class_="product_pod")
    except Exception as e:
        print(f"Error loading page {i}: {e}")
        continue
        
    for contant in contants:
        # Name
        try:
            name.append(contant.find("h3").find("a")['title'])
        except:
            name.append(None)

        # Price
        try:
            price_text = contant.find("div", class_="product_price").find("p").text
            price.append(price_text.split("£")[1])
        except:
            price.append(None)

        # Rating
        try:
            rating.append(contant.find("p", class_='star-rating')['class'][1])
        except:
            rating.append(None)

        # Availability
        try:
            avilablity.append(contant.find("p", class_="instock availability").text.strip())
        except:
            avilablity.append(None)

        # Book link
        try:
            link = contant.find("div", class_="image_container").find("a")["href"]
            book_info_link.append(link)
        except:
            book_info_link.append(None)
data = {"Name":name,"Price":price,"Rating":rating,"Avilablity":avilablity,"book_link":book_info_link}
book_df = pd.DataFrame(data)
```

```python
book_df.to_csv("Book Data Scrape.csv")
```
